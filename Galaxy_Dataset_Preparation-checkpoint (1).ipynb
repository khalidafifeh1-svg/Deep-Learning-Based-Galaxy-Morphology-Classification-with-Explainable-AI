Organize Images by Class

```python
import os
import shutil
import re
from tqdm import tqdm  # progress bar

# Path where raw images
raw_images_dir = r"C:\Users\khali\Synoptic Project\images_gz2\images"
# Output folder for organized dataset
output_dir = r"C:\Users\khali\Synoptic Project\dataset_gz2"
# Function to remove illegal characters from folder names
def sanitize(name):
    return re.sub(r'[\\/*?:"<>|]', "_", name)

# Get all unique galaxy classes
classes = merged_df['gz2_class'].unique()
# Create a folder for each class
for cls in classes:
    os.makedirs(os.path.join(output_dir, sanitize(cls)), exist_ok=True)
# Loop through dataset rows one by one
for _, row in tqdm(merged_df.iterrows(), total=len(merged_df), desc="Copying images"):
    asset_id = str(row['asset_id']) # image ID
    label = sanitize(row['gz2_class']) # class label

    found = False # track if image is found
    # Try different image extensions
    for ext in ['.jpg', '.jpeg', '.png']: 
                # Build filename
        filename = f"{asset_id}{ext}"
          # Full path to image
        src = os.path.join(raw_images_dir, filename)
        # If image exists
        if os.path.exists(src):
         # Destination path
            dst = os.path.join(output_dir, label, filename)
                        # Copy image to class folder
            shutil.copy2(src, dst)
            found = True
            break
    # If image not found print warning

    if not found:
        tqdm.write(f"Warning: {asset_id} not found")
```

Main Classes

```python
import os
import shutil
from tqdm import tqdm  # progress bar

# Input folder 
source_dir = r"C:\Users\khali\Synoptic Project\dataset_gz2"
# Output folder
target_dir = r"C:\Users\khali\Synoptic Project\dataset_gz2_main"
# Function to map detailed labels into main categories
def map_to_main_class(label):
     # If label starts with SB it go to Barred Spiral
    if label.startswith('SB'):
        return 'Barred_Spiral'
            # If label starts with E it go to Elliptical galaxy
    elif label.startswith('E'):
        return 'Elliptical'
            # If label starts with S it go to Spiral galaxy
    elif label.startswith('S'):
        return 'Spiral'
            # Otherwise Irregular type
    else:
        return 'Irregular'
#  main class folders
main_classes = ['Barred_Spiral', 'Elliptical', 'Spiral']
# Create folders for main classes
for cls in main_classes:
    os.makedirs(os.path.join(target_dir, cls), exist_ok=True)

# Count total files for progress bar
total_files = 0
    # Only count folders
for folder in os.listdir(source_dir):
    src_path = os.path.join(source_dir, folder)
    if os.path.isdir(src_path):
        total_files += len(os.listdir(src_path))

# Progress bar while moving files
with tqdm(total=total_files, desc="Moving files") as pbar:
# Loop through each subclass folder
    for folder in sorted(os.listdir(source_dir)):
        src_path = os.path.join(source_dir, folder)

        if not os.path.isdir(src_path):
            continue
        # Get main class name
        main_class = map_to_main_class(folder)
                # Destination folder
        dst_folder = os.path.join(target_dir, main_class)
        # Loop through images in folder
        for file in sorted(os.listdir(src_path)):
            src_file = os.path.join(src_path, file)

            if os.path.isfile(src_file):

                dst_file = os.path.join(dst_folder, file)

                # Avoid duplicate filenames
                counter = 1
                base_name, ext = os.path.splitext(file)

                while os.path.exists(dst_file):
                    dst_file = os.path.join(
                        dst_folder,
                        f"{base_name}_copy{counter}{ext}"
                    )
                    counter += 1

                shutil.move(src_file, dst_file)

                pbar.update(1)

print(" Main class organization complete.")
```

SPLIT

```python
import os
import shutil
import random
from tqdm import tqdm

# dataset folder
source_dir = r"C:\Users\khali\Synoptic Project\dataset_gz2V2"
# Output split dataset folder
output_dir = r"C:\Users\khali\Synoptic Project\dataset_split2"

# Ratios for splitting data
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# Define dataset splits
splits = ['train', 'val', 'test']
# Create folder structure
for split in splits:
    for class_name in os.listdir(source_dir):
        class_path = os.path.join(source_dir, class_name)
        if os.path.isdir(class_path):
            os.makedirs(os.path.join(output_dir, split, class_name), exist_ok=True)

# Loop through each class
for class_name in os.listdir(source_dir):
    class_path = os.path.join(source_dir, class_name)

    if not os.path.isdir(class_path):
        continue
    # Get all images
    images = os.listdir(class_path)
    random.shuffle(images)
    # calculate split indices
    total = len(images)
    train_end = int(train_ratio * total)
    val_end = int((train_ratio + val_ratio) * total)
    # Split data
    train_imgs = images[:train_end]
    val_imgs = images[train_end:val_end]
    test_imgs = images[val_end:]

    # Copy training images
    for img in tqdm(train_imgs, desc=f"{class_name} - Train"):
        shutil.copy2(
            os.path.join(class_path, img),
            os.path.join(output_dir, 'train', class_name, img)
        )
    # Copy validation images
    for img in tqdm(val_imgs, desc=f"{class_name} - Val"):
        shutil.copy2(
            os.path.join(class_path, img),
            os.path.join(output_dir, 'val', class_name, img)
        )
    # Copy test images
    for img in tqdm(test_imgs, desc=f"{class_name} - Test"):
        shutil.copy2(
            os.path.join(class_path, img),
            os.path.join(output_dir, 'test', class_name, img)
        )

print("\n Dataset successfully split!")
print(" Output folder:", output_dir)
```

Clean Corrupted Images

```python
from PIL import Image
import os
from tqdm import tqdm

# Function to clean dataset
def clean_dataset(root_dir, name="dataset"):

    broken_files = []  # store corrupted images
    total_files = 0

    # Count all images
    for folder in os.listdir(root_dir):
        folder_path = os.path.join(root_dir, folder)

        if os.path.isdir(folder_path):
            total_files += len(os.listdir(folder_path))

    print(f"Checking {name}: {total_files} images")

    # Progress bar
    with tqdm(total=total_files, desc=f"Cleaning {name}") as pbar:

        for folder in os.listdir(root_dir):

            folder_path = os.path.join(root_dir, folder)

            if not os.path.isdir(folder_path):
                continue

            for file in os.listdir(folder_path):

                file_path = os.path.join(folder_path, file)

                try:
                    img = Image.open(file_path)
                    img.verify()  # check if image is valid

                except:
                    print(f"Removing broken image: {file_path}")
                    broken_files.append(file_path)
                    os.remove(file_path)

                pbar.update(1)

    print(f"{name} cleaning complete")
    print(f"Removed {len(broken_files)} images")

    return len(broken_files)

train_dir = r"C:\Users\khali\Synoptic Project\dataset_split\train"
val_dir   = r"C:\Users\khali\Synoptic Project\dataset_split\val"
test_dir  = r"C:\Users\khali\Synoptic Project\dataset_split\test"

print("STARTING DATASET CLEANING...\n")

train_removed = clean_dataset(train_dir, "TRAIN")
val_removed   = clean_dataset(val_dir, "VAL")
test_removed  = clean_dataset(test_dir, "TEST")

print("FINAL SUMMARY")
print("Train removed:", train_removed)
print("Val removed:", val_removed)
print("Test removed:", test_removed)
```

Dataset Balancing and  Augmentation

```python
import os  # Used for file/folder operations (create, list directories)
import random  
import shutil  # Used to copy files from one folder to another
from PIL import Image 
from torchvision import transforms  
from tqdm import tqdm 

# Path to training dataset (input images grouped by class)
input_dir = r"C:\Users\khali\Synoptic Project\dataset_split\train"

# Path where balanced dataset will be saved
output_dir = r"C:\Users\khali\Synoptic Project\dataset_balanced\train"

# Target number of images per class 
target_size = 67000  

# Define augmentation pipeline 
augment = transforms.Compose([
    transforms.RandomHorizontalFlip(),  # Flip image left-right randomly
    transforms.RandomRotation(20),  # Rotate image randomly up to ±20 degrees
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),  # Crop + resize image
])

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through each class folder 
for class_name in os.listdir(input_dir):

    # Full path to current class folder
    class_path = os.path.join(input_dir, class_name)

    # Output folder for this class
    output_class_path = os.path.join(output_dir, class_name)

    # Skip if not a folder 
    if not os.path.isdir(class_path):
        continue

    # Create output class folder
    os.makedirs(output_class_path, exist_ok=True)

    # Get all image files in this class folder
    images = [
        f for f in os.listdir(class_path)
        if f.lower().endswith(('.png', '.jpg', '.jpeg'))  # only images
    ]

    # Count how many images exist in this class
    current_count = len(images)

    # Print class information
    print(f"\n {class_name}: {current_count}")

   
    if current_count > target_size:

        # Randomly pick only target_size images
        selected = random.sample(images, target_size)

        # Copy selected images to new folder
        for img_name in tqdm(selected, desc=f"{class_name} (copying)"):

            src = os.path.join(class_path, img_name)  # source image path
            dst = os.path.join(output_class_path, img_name)  # destination

            shutil.copy2(src, dst)  # copy image file


    else:

        # First copy ALL original images
        for img_name in tqdm(images, desc=f"{class_name} (copying originals)"):

            src = os.path.join(class_path, img_name)
            dst = os.path.join(output_class_path, img_name)

            shutil.copy2(src, dst)

        # Calculate how many new images we need to create
        needed = target_size - current_count

        print(f"Augmenting {needed} images for {class_name}")

        i = 0  # counter for new images

        # Progress bar for augmentation process
        pbar = tqdm(total=needed, desc=f"{class_name} (augmenting)")

        # Keep generating images until we reach target size
        while i < needed:

            # Randomly pick an existing image
            img_name = random.choice(images)

            # Full image path
            img_path = os.path.join(class_path, img_name)

            try:
                # Open image and convert to RGB format
                img = Image.open(img_path).convert("RGB")

                # Apply random augmentation transformations
                aug_img = augment(img)

                # Create new unique filename
                new_name = f"aug_{i}_{img_name}"

                # Save augmented image
                aug_img.save(os.path.join(output_class_path, new_name))

                # Increase counter
                i += 1

                # Update progress bar
                pbar.update(1)

            except:
                # If image fails  skip it
                continue

        # Close progress bar
        pbar.close()

# Final message
print("\n Dataset balanced successfully!")
```
